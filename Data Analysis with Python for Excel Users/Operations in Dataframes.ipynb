```python
# Operations in rows
# Adding a Total Score and Average Score columns to the dataframe
df_exams['Total Score'] = df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
df_exams['Avg. Score'] = df_exams['Total Score']/3
df_exams['Avg. Score']
df_exams
```

## Sorting the columns

```python
# Sort by one column
df_exams.sort_values(by='math score')
```

```python
# Sort by descending by one column
df_exams.sort_values('math score', ascending=False)
```

```python
# Sort by muliple columns
df_exams.sort_values(['math score', 'reading score'], ascending=False)
```

```python
# Update dataframe with sorting - default is that the sorting doesn't affect the original dataframe
df_exams_sorted = df_exams.sort_values(['math score', 'reading score'], ascending=False)
```
